## Usage

### HLTVMatch( )
Provide team_url in such format "/matches/2361342/natus-vincere-vs-outsiders-iem-katowice-2023"

```python
from hltv_stats import HLTVMatch
match_url = "/matches/2361342/natus-vincere-vs-outsiders-iem-katowice-2023"
match = HLTVMatch(match_url)
```

```python
match.parse_analytics_summary(filename=None)
```

```python
match.parse_head_to_head()
```

```python
match.parse_pick_ban_stats()
```

```python
match.parse_analytics_center()
```

### HLTVTeam( )
Explicitly provide team url, or use HLTVMatch().teams_link, 
which is a list of team's urls ['/4608/natus-vincere', '/11595/outsiders'] 

```python
match.teams_link
```

```python
from hltv_stats import HLTVTeam
team1 = HLTVTeam("/4608/natus-vincere") #or "match.teams_link[0], match.teams_link[1]"
```

```python
#You can use match_id to assign team current state(statistic) to specific match
team1.match_id = match.match_id
```

```python
team1.parse_matches(time_filter=3, filename=None)
```

```python
team1.parse_players(time_filter=3, filename=None)
```

```python
team1.parse_maps(time_filter=3, filename=None)
```

```python
team1.parse_events(time_filter=3, filename=None)
```

```python
#returns tuple of all fucntions above
team1.parse_all_stats(time_filter=1) 
```
